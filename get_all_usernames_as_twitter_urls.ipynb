```r
# load and optionally install required packages
if (!require('pacman')) install.packages('pacman')
pacman::p_load(
    dplyr,
    readr
)
```

```r
# initial read in of all data
csv_files <- list.files(path='./data/raw/talkwalker', pattern='*.csv', full.names = TRUE)
data_groups <- lapply(csv_files, read_csv)
data <- do.call(rbind, data_groups)
```

```r
urls <- data %>% select(-url) %>% rename(url = extra_author_attributes.url) %>% distinct(url) %>% select(url)
```

```r
urls %>% write_csv('data/generated/urls.csv')
```
