```r
# initial read in of all data
data.raw <- read_csv('./data/generated/data_with_tokenized_distinct.csv')
keywords.raw <- read_csv('./data/generated/cluster_keywords_distinct.csv')
```

```r
keywords <- keywords.raw %>%
mutate(
    keywords = str_split(keywords, ' ')
)
data <- data.raw %>%
mutate(
    tokens = str_split(tokenized, ' ')
)
```

```r
keywords
```

```r
data.guessed <- data %>%
merge(keywords) %>%
rowwise() %>%
mutate(
    matched_keywords = intersect(as.list(tokens), as.list(keywords)) %>% paste(collapse=','),
    matched_keywords_count = intersect(as.list(tokens), as.list(keywords)) %>% length()
) %>%
filter(matched_keywords_count > 0) %>%
group_by(content) %>%
top_n(1, matched_keywords_count) %>%
arrange(desc(matched_keywords_count)) %>%
select(content, keywords, tokens, matched_keywords, matched_keywords_count, cluster_number, published) %>%
mutate(
    id = cluster_number
)
data.guessed
```

```r
write_to_file = function(data.to_write){
    write_csv(data.to_write, paste0('./data/generated/tweets_by_cluster_distinct_', unique(data.to_write$id), '.csv'))
    write(data.to_write %>% toJSON(), paste0('./data/generated/tweets_by_cluster_distinct_', unique(data.to_write$id), '.json'))
    return(data.to_write)
}

keywords %>%
mutate(
    id = cluster_number
) %>% merge(data.guessed) %>%
group_by(id) %>%
do(write_to_file(.))
```
