```r
# load and optionally install required packages
if (!require('pacman')) install.packages('pacman')
pacman::p_load(
    dplyr,
    readr,
    scales
)
```

```r
data <- read_csv('./data/generated/debotted_combined.csv')
```

```r
data %>% filter(grepl('^RT : .*', content))
```

```r
data.removed_rts <- data %>% filter(!grepl('^RT : .*', content)) %>% distinct(content)
```

```r
data.removed_rts %>% nrow()
```

```r
data.count <- data %>% nrow()
data.removed_rts.count <- data.removed_rts %>% nrow()
percent(data.removed_rts.count / data.count)
```

```r
data.removed_rts %>% write_csv('./data/generated/debotted_combined_no_rts.csv')
```
